```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('minimum_wage_PD.xlsx')
data
```

```python
data = data[data['Year'] >= 1990]
data.head(5)
```

```python
data['Monthly Minimum Wage'] = data['Real Minimum Wage'] / 12
```

```python
plt.figure(figsize=(20,10))
grouped = data.groupby('Country')
for name, group in grouped:
    plt.plot(group['Year'], group['Monthly Minimum Wage'], label=name)

plt.legend(loc='lower right')
plt.xlabel('Year')
plt.ylabel('Monthly Real Minimum Wage')
plt.grid()
plt.show()
```
